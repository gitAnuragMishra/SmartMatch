```python
def extract_skills_from_resume(text, skills_list):
    skills = []

    for skill in skills_list:
        pattern = r"\b{}\b".format(re.escape(skill))
        match = re.search(pattern, text, re.IGNORECASE)
        if match:
            skills.append(skill)

    return skills
```

```python
with open("extractor_library.json", "r") as file:
    education_data = json.load(file)

education_keywords = education_data["education_keywords"]
```

```python
def extract_education_from_resume(text, education_keywords):
    education = []

    for keyword in education_keywords:
        pattern = r"(?i)\b{}\b".format(re.escape(keyword))
        match = re.search(pattern, text)
        if match:
            education.append(match.group())

    return education
```

```python
extracted_education = extract_education_from_resume(resume, education_keywords)

# Display results
if extracted_education:
    print("Education:", extracted_education)
else:
    print("No education information found.")
```
